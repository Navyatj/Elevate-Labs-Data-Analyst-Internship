### DATA ANALYST INTERNSHIP 

#### Task 1: Data Cleaning and Preprocessing

##### Objective: Clean and prepare a raw dataset (with nulls, duplicates, inconsistent formats)

```python
import pandas as pd 
#loading the dataset and examining the data
df = pd.read_csv('Medical Appoinment No shows.csv', encoding='ISO-8859-1')
df.head()
```

##### checking for missing values

```python
# Check for the total count of missing values in each column
missing_values = df.isnull().sum()
print(missing_values)
```

##### Standardizing the column Headers

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
# Rename specific columns with misspellings or inconsistent naming
df.rename(columns={
    'patientid': 'patient_id',
    'appointmentid': 'appointment_id',
    'scheduledday': 'scheduled_day',
    'appointmentday': 'appointment_day',
    'hipertension': 'hypertension',
    'handcap': 'handicap',
    'sms_received': 'sms_received',
    'no-show': 'no_show',
    'alcoholism': 'alcoholism'
}, inplace=True)
```

```python
df.head()
```

##### Convert date formats to a consistent type

```python
# Convert the columns to datetime objects
df['scheduled_day'] = pd.to_datetime(df['scheduled_day'])
df['appointment_day'] = pd.to_datetime(df['appointment_day'])
```

```python
# Format the datetime objects as 'dd-mm-yyyy' strings
df['scheduled_day'] = df['scheduled_day'].dt.strftime('%d-%m-%Y')
df['appointment_day'] = df['appointment_day'].dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

##### Check and fix data types

```python
#Examining the datatypes of data
df.info()
```

PatientId is float,ScheduledDay and AppointmentDay is object This should be converted into int and datetime respectively

```python
# Convert the 'patient_id' column to an integer data type
df['patient_id'] = df['patient_id'].astype(int)
```

```python
df.info()
```

```python
import pandas as pd

# The format of the dates is day-month-year
date_format = '%d-%m-%Y'

# Convert the 'appointment_day' column
df['appointment_day'] = pd.to_datetime(df['appointment_day'], format=date_format)

# Convert the 'scheduled_day' column
df['scheduled_day'] = pd.to_datetime(df['scheduled_day'], format=date_format)
```

```python
df.info()
```

##### Standardize text values like gender

```python
gender_mapping = {
    'F': 'Female',
    'M': 'Male'
}
```

```python
df['gender'] = df['gender'].replace(gender_mapping)
```
